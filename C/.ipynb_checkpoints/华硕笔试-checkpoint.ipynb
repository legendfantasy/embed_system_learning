7. 阅读下列代码，选项中描述正确的是（ ）

```python
typedef struct LNode
{
    int data;
    struct LNode *next;
};
//参数 head 是一线性链表的头指针，若链表有结尾，结尾指针为NULL
bool a(LNode *head)
{
    LNode *slow = head, *fast = head;
    while (fast && slow)
    {
        slow = slow->next;
        if (fast->next == NULL)
            fast = fast->next;
        else
            fast = fast->next->next;
        if (slow == fast)
            break;
    }
    return !(fast == NULL);
}
bool b(LNode *head)
{
    LNode *t_head = head;
    if(head == NULL)
        return false;
    LNode *p1 = head;
    LNode *p2 = head->next;
    LNode *p3 = NULL;
    t_head->next = NULL;
    while(p2 != NULL)
    {
        if(p2 == t_head)
            return true;
        p3 = p2->next;
        p2->next = p1;
        p1 = p2;
        p2 = p3;
    }
    return false;
}
```

A. 函数a与b在输入链表相同的情况下返回值一定相同  
B. 函数a与b在输入链表相同的情况下返回值一定不相同  
C. 函数b没有破坏链表的结构  
D. 函数a破坏了链表的结构

22. 下列代码输出内容（）

```python
#include <stdio.h>
int inc(int a)
{
    return (++a);
}
int multi(int *a, int *b, int *c)
{
    return (*c = *a++**b+++*a);
}
typedef int(*FUNC1)(int in);  
typedef int(*FUNC2)(int*, int*, int*);  
void show(FUNC2 fun, int arg1, int *arg2)
{
    FUNC1 p = &inc;  
    int temp = p(arg1);
    fun(&temp, &arg1, arg2);
    printf("%d\n", *arg2);
}
int main()
{
    int a;
    show(multi, 9, &a);
    return 0;
}
```

A. 100  
B. 110  
C. 101  
D. 90
