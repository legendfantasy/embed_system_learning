# Internet历史  
1957年,苏联先后有两颗卫星上天,1958年美国总统艾森豪威尔向美国国会提出建立DARPA(Defense Advanced Research Project Agency),即国防部高级计划研究署,简称ARPA  

1968年,这个机构提出资源共享计算机网络(Resource Sharing Computer Networks),让这个国防部的下属机构所有的电脑都互连起来,这个网络叫做ARPANet,是现代互联网的前身  

早期的ARPA网络使用网络控制协议(Network Control Protocol,NCP)不能互联不同类型的操作系统,也没有纠错功能   

1973年,Robert Kahn 和 Vinton Cerf两人合作开发了新的互联网协议.1974年两人正式发表第一份TCP(Transmission Control Protocol)协议的详细说明  

随着互联网的发展,这份协议发展成了今天的TCP/IP协议    
TCP协议分成了两个不同的协议:    
+ 用来检测网络传输中差错的传输控制协议TCP(可靠传输)  
+ 专门负责对不同网络进行互联的互联网协议IP(不可靠传输)    

1983年ARPANet停止服务,至此互联网上的主机全部使用TCP/IP协议

# 网络的体系结构:
网络采用分而治之的方法设计,将不同的功能划分为不同的模块,以分层形式有机组合在一起  
每层实现不同的功能,其内部实现方法对外部其他层次来说都是透明的.(进行内部细节封装,对外不可见)  

网络的体系结构即指网络层次结构和每层所使用协议的集合    

目前有两类非常重要的体系结构:OSI与TCP/IP   

## OSI 7层模型  
这是官方发布的一个比较理想的网络体系结构,比较通用却很少使用因此没有完整实现  
 
    
和这个联系紧密的一个概念是交换机,通常所说的二层交换机指用于数据链路层的交换机,三层交换机是指用于网络层的交换机   
## TCP/IP 体系结构  
这个网络体系是事实上使用的工业标准    
    
+ 网络接口和物理层:用于屏蔽不同网络连接方式(Wifi、GPRS、3G、4G、光纤等）的硬件差异，linux驱动到网络层会提供一个统一的接口    
+ 网络层：实现端到端的传输（IP协议ICMP协议IGMP协议）    
+ 传输层：（TCP和UDP）决定数据应当交给哪一个任务来处理    
+ 应用层： HTTPS的S代表secret     

每层协议：   
  
网络接口与物理层：    
ARP（地址解析协议）：根据IP地址查询到对应的MAC地址    
MAC地址是每个联网设备分配的唯一标志    
RARP：根据MAC地址找到对应的IP地址    
ppp拨号协议：ADSL，GPRS/3G/4G的协议    
网络层（IP层）：    
IP： Internet protocol 分为IPV4和IPV6    
ICMP：Internet控制管理协议，ping命令属于ICMP   
IGMP：Internet分组管理协议，广播和组播    
传输层：  
TCP：Transfer Control protocol 传输控制协议 提供面向连接的，一对一的可靠数据传输协议  
UDP：User Datagram Protocol 用户数据报协议 提供不可靠，无连接的尽力传输协议  
SCTP：Stream Control Transmission Protocol 流控制传输协议 TCP的增强版 提供面向连接的 多主机，多链路通信  
应用层：  
网页访问协议；POP3（收）/SMTP(发) IMAP(可接收邮件的一部分)  
FTP  
Telnet/SSH 远程登陆  
嵌入式相关：  
NTP：网络时钟协议  
SNMP：简单网络管理协议，实现网络设备集中式管理  
RTP/RTSP:用传输音视频的协议（安防监控）

# socket
+ 是一个应用编程接口
+ 是一种特殊的文件描述符，可以对其圣是IO操作函数（read、write、close）
+ 并不限于TCP/IP协议
+ 面向连接（TCP/IP)
+ 无连接（UDP、IPX（Inter-network Packet Exchange)
+ 代表着网络编程的一种资源

Socket类型：
+ 流式套接字（SOCK_STREAM)
  提供了一个面向连接、可靠的数据传输报务。数据无差错、无重复的发送且按发送顺序接收，内设置流量控制，避免数据流淹没慢的接收方，数据被看作是字节流，无长度限制  
  唯一对应TCP  
+ 数据报套接字（SOCK_DGRAM）
  提供无连接服务，数据包以独立数据包的形式被发送，不提供无差错保证，数据可能丢失或重复，顺序发送，可能被乱序接收  
  唯一对应UDP  
+ 原始套接字（SOCK_RAW）
  可以对较低层次协议如IP、ICMP直接访问

# socket 套接字

网络编程接口(API)  
socket 常用API:  
```
/*创建套接字*/
int socket(int domain, int type, int protocol);
/*绑定通信结构体*/
int bind(int sockfd, const, struct sockaddr *addr, socklen_t addrlen);
/*监听套接字*/
int listen(int sockfd, int backlog);
/*处理客户端发起的连接，生成新的套接字*/
int accept(int sockfd, struct sockaddr *addr, socklen_t addrlrn);
/*向服务器发起连接请求*/
int connect(int sockfd, const struct sockaddr *addr, socklen_t addrlrn);
```

三元组:  
+ IP地址: 标识计算机  
+ 端口: 标识计算机中的当前进程  
+ 协议: 指定数据传输的方式

## 服务器端

````c
#include <stdio.h>
#include <stdlib.h>
#include <unistd.h>
#include <sys/socket.h>
#include <sys/types.h>
#include <arpa/inet.h>
#include <string.h>
 
#define BACKLOG 5
 
int main(int argc, char *argv[])
{
	int fd, newfd, ret; //文件描述符 
	char buf[BUFSIZ] = {}; //BUFSIZ 8142字节
	struct sockaddr_in addr;//结构体负责存储iP地址和端口号

	if (argc < 3)           //提示传入的参数个数
    {
		fprintf(stderr, "%s<addr><port>\n", argv[0]);
		exit(0);
	}
 
	/*创建套接字 */
	fd = socket(AF_INET, SOCK_STREAM, 0);// 
	if (fd < 0ss
    {
		perror("socket");
		exit(0);
	}
    //设置 addr 结构体
	addr.sin_family = AF_INET;
	addr.sin_port = htons(atoi(argv[2]));  //参数是端口号
	if (inet_aton(argv[1], &addr.sin_addr) == 0) //检查ip地址转化是否成功
    {
		fprintf(stderr, "Invalid address\n");  //函数不在perror中留下信息
		exit(EXIT_FAILURE);
	}
 
	/*绑定通信结构体*/
	if(bind(fd, (struct sockaddr *)&addr, sizeof(addr)) == -1)
    {
		perror("bind");
		exit(0);
	}
 
	/*设置套接字为监听模式*/
	if (listen(fd, BACKLOG) == -1) 
    {
		perror("listen");
		exit(0);
	}
 
 
	/*接受客户端的连接请求,生成新的用于和客户端通信的套接字*/
	newfd = accept(fd, NULL, NULL);
	if (newfd < 0) 
    {
		perror("newfd");
		exit(0);
	}
 
	while(1) 
    {
		memset(buf, 0, BUFSIZ);
		ret = read(newfd, buf, BUFSIZ);
		if(ret < 0) 
        {
			perror("read");
			exit(0);
		}else if(ret == 0)
			break;
		else
			printf("buf = %s\n", buf);
	}
 
	close(newfd);
	close(fd);
 
	return 0;
 
}
````

````c
#include <stdio.h>
#include<sys/socket.h>
#include <sys/types.h>
#include <stdlib.h>
#include <arpa/inet.h>
#include <unistd.h>
#include <string.h>
 
#define BACKLOG 5
#define STR "hello world"
 
int main (int argc, char *argv[]) 
{
	int fd;
	struct sockaddr_in addr;
	char buf[BUFSIZ] = {};
 
	if (argc < 3)
    {
		fprintf(stderr, "%s<addr><port>\n", argv[0]);
		exit(0);
	}
 
	/*创建套接字 */
	fd = socket(AF_INET, SOCK_STREAM, 0);
	if (fd < 0) 
    {
		perror("socket");
		exit(0);
	}
	addr.sin_family = AF_INET;
	addr.sin_port = htons(atoi(argv[2]));
	if (inet_aton(argv[1], &addr.sin_addr) == 0) 
    {
		fprintf(stderr, "Invalid address\n");
		exit(EXIT_FAILURE);
	}
	/*向服务端发起连接请求*/
	if(connect(fd, (struct sockaddr *)&addr, sizeof(addr)) == -1) 
    {
		perror("connect");
		exit(0);
	}
	while(1) 
    {
		printf("input->");
		fgets(buf, BUFSIZ, stdin);
		write(fd, buf, strlen(buf));
	}
	close(fd);
 
 
	return 0;
}
````
