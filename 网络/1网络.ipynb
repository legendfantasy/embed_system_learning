# Internet 历史



1957 年,苏联先后有两颗卫星上天,1958 年美国总统艾森豪威尔向美国国会提出建立 DARPA(Defense Advanced Research Project Agency),即国防部高级计划研究署,简称 ARPA

1968 年,这个机构提出资源共享计算机网络(Resource Sharing Computer Networks),让这个国防部的下属机构所有的电脑都互连起来,这个网络叫做 ARPANet,是现代互联网的前身

早期的 ARPA 网络使用网络控制协议(Network Control Protocol,NCP)不能互联不同类型的操作系统,也没有纠错功能

1973 年,Robert Kahn 和 Vinton Cerf 两人合作开发了新的互联网协议.1974 年两人正式发表第一份 TCP(Transmission Control Protocol)协议的详细说明

随着互联网的发展,这份协议发展成了今天的 TCP/IP 协议  
TCP 协议分成了两个不同的协议:

- 用来检测网络传输中差错的传输控制协议 TCP(可靠传输)
- 专门负责对不同网络进行互联的互联网协议 IP(不可靠传输)

1983 年 ARPANet 停止服务,至此互联网上的主机全部使用 TCP/IP 协议

# 网络的体系结构:



网络采用分而治之的方法设计,将不同的功能划分为不同的模块,以分层形式有机组合在一起

网络的体系结构即指网络层次结构和每层所使用协议的集合

目前有两类非常重要的体系结构:OSI 与 TCP/IP

每层实现不同的功能,其内部实现方法对外部其他层次来说都是透明的.(进行内部细节封装,对外不可见)，
每一层都向上一层级屏蔽这一层的实现细节，中间过程就好像在这一层全部实现

所以网络层级的划分要考虑的是将网络的功能分成哪些部分，哪些部分又可以进行抽象概括

## 五层体系结构



- 应用层
  - 任务：通过应用进程间的交互来完成特定网络应用。
  - 协议：定义的是应用进程间通信和交互的规则。
  - 把应用层交互的数据单元称为报文(message)。
  - 例如：DNS，HTTP，SMTP
- 运输层
  - 任务：负责向两台主机中进程之间的通信提供通用的数据传输服务。
  - 具有复用和分用的功能。
  - 主要使用两种协议：  
    + 传输控制协议 TCP 
      提供面向连接的、可靠的数据传输服务 数据传输的单位是报文段 (segment）  
    + 用户数据报协议 UDP 
      提供无连接的尽最大努力 (best-effort) 的数据传输服务（不保证数据传输的可靠性） 数据传输的单位是用户数据报
      ![Alt text](image-33.png)
- 网络层
  - 为分组交换网上的不同主机提供通信服务。
  - 两个具体任务：
    - 路由选择：通过一定的算法，在互联网中的每一个路由器上，生成一个用来转发分组的转发表。
    - 转发：每一个路由器在接收到一个分组时，要依据转发表中指明的路径把分组转发到下一个路由器
  - 互联网使用的网络层协议是无连接的网际协议 IP (Internet Protocol) 和许多种路由选择协议，因此互联网的网络层也叫做网际层或 IP 层 IP 协议分组也叫做 IP 数据报，或简称为数据报。
- 数据链路层
  - 常简称为链路层。
  - 任务：实现两个相邻节点之间的可靠通信 在两个相邻节点间的链路上传送帧（frame）如发现有差错，就简单地丢弃出错帧 如果需要改正出现的差错，就要采用可靠传输协议来纠正出现的差错。这种方法会使数据链路层协议复杂
- 物理层
  - 任务：实现比特（0 或 1）的传输。
  - 确定连接电缆的插头应当有多少根引脚，以及各引脚应如何连接。
  - 注意：传递信息所利用的一些物理媒体，如双绞线、同轴电缆、光缆、无线信道等，并不在物理层协议之内，而是在物理层协议的下面。

# Mac地址
## Mac地址的含义
1. Mac地址又称物理地址，用来标志每个设备在网络中的位置，不同于为了方便管理而设置的软件意义上的地址，MAC地址属于物理意义上的地址，对每个设备而言都是唯一的  
2. MAC 地址的英语是 Media Access Control Address，直译为媒体存取控制位址，也称为局域网地址（LAN Address）、以太网地址（Ethernet Address）或物理地址（Physical Address）  
## Mac地址的表示方法
   MAC 地址采用十六进制数表示，共 6 个字节（48 位），长度为 48bit（字节）。整个地址可以分为前 24 位和后 24 位，代表不同的含义。前 24 位称为组织唯一标识符（Organizationally UniqueIdentifier，OUI），是由 IEEE 的注册管理机构给不同厂家分配的代码，区分了不同的厂家。后 24 位是由厂家自己分配的，称为扩展标识符。同一个厂家生产的网卡中 MAC 地址后 24 位是不同的。  
## Mac地址的局限性/为什么不直接用Mac地址作为网络通讯的地址
4. Mac地址只能做到标记硬件设备，只能指示生产厂商和生产地址，不能在地址空间上引入逻辑结构，使其无法具备真正的地址来表示国家、省、市、区、街道、路、号这类层次。
5. 因此，要进行数据传输，必须使用一种逻辑化、层次化的寻址方案对网络进行组织，这就是 IP 地址

1. 采用2级结构划分  
   
   ![alt text](2级结构划分.png)
   
   网络部分标识设备所在的特定网络，而主机部分标识该网络内的特定设备（或主机）  
   
   让我们以一个典型的IPv4地址为例：`192.168.1.100`
   在这个IP地址中：
   - `192.168.1`是网络部分。它标识设备所在的特定网络。这个网络上的所有设备的IP地址都会以`192.168.1`开始。
   - `100`是主机部分。它标识`192.168.1`网络内的特定设备（或主机）。这个网络上的每个设备都会有一个唯一的主机部分。
   所以，这个IP地址的二级结构允许数据被路由到正确的网络（`192.168.1`），然后路由到该网络上的正确设备（`100`）

# 端口号



## 端口号的作用：
  为了区分一台主机接收到的数据包应该交给哪个任务来进行处理  
  属于运输层的功能实现  
  
  ![alt text](端口号的作用.png)

## 端口号的构成
- 使用 16 位数字（1-65535）
- 端口号一般由 IANA (Internet Assigned Numbers Authority) 管理

## 端口号的划分
- 众所周知端口：1 ∼ 1023（1 ∼ 255 之间为众所周知端口，256~1023 端口通常由 UNIX 系统占用。
    FTP: 21，SSH： 22， HTTP:80， HTTPS:469）
- 注册端口（保留端口）：1024~49150
- 动态或私有端口：49151~65535
- 保留端口： 1024-5000（不建议使用）
- 可以使用的：5000~65535
- TCP 端口号与 UDP 端口号独立

  ![alt text](端口号的划分.png)

## BSD端口：
+ 0不使用，1-1023为系统端口，也叫BSD保留端口。
+ 0-1023： BSD保留端口，也叫系统端口，这些端口只有系统特许的进程才能使用
+ 1024~65535为用户端口，其中：
    1024-5000： BSD临时端口，一般的应用程序使用1024到4999来进行通讯；  
    5001-65535：BSD服务器(非特权)端口，用来给用户自定义端口。  
+ IANA建议49152至65535作为“动态或私有端口”。
+ 许多Linux内核使用32768至61000范围。配置文件/proc/sys/net/ipv4/ip_local_port_range有当前系统设定。

# 字节序


## 主机字节序(HBO - Host Byte Order) 

不同类型 CPU 的主机中，内存**存储**多字节整数序列有两种方法，称为主机字节序(HBO)  

> 大端字节序和小端字节序

计算机电路在处理低位字节时，效率比较高，因为计算机都是从低位开始的，所以，计算机的内部处理大多都是小端字节序；

但是在程序员的眼里，还是习惯读写大端字节序，所以除了计算机内部外，其他场合大都使用大端字节序；如网络传输和文件存储中；

字节序是指不同的 CPU 访问内存中的多字节数据的时候，存在大小端的问题，如果 CPU 访问的是字符串，则不存在大小端的问题

- 小端：低对低 低端内存存放低端数据
- 大端：低对高 低端内存存放高端数据  

![alt text](大端对齐和小端对齐.png)

![alt text](大端字节序和小端字节序.png)

- 一般来说：  
  X86/ARM:小端  
  powerpc/mips, ARM 作为路由器时：大端
- 网络传输的时候采用大端模式  

## 网络字节序(NBO-Network Byte Order)

为了避免不同类别主机之间在数据交换时由于对于字节序的不同而导致的差错，引入了网络字节序。 网络字节序统一采用大端字节序，网络编程中将本地字节序和网络字节序进行转换

## 字节序的转换 

在大部分 PC 机上，当应用进程将整数送入 socket 前，需要转化成网络字节序；  
当应用进程从 socket 取出整数后，要转化成小端字节序

### IP 地址的转换

inet 系列函数

1. inet_aton()

```c
#include <arpa/inet.h>
int inet_aton(const char *strptr, struct in_addr *addrptr);  //adress to numeric
```

- 功能：将 strptr 所指的 IPv4 点分十进制表示法转换成 32 位的网络字节序二进制值，并将结果存储在 addrptr 指向的结构中。
- 参数：
  - strptr ： IPv4 点分十进制 IP 地址
  - addrptr : 结构体指针，用于存储转化后的 32 位网络地址
- 返回值：
  - 如果地址转换有效，返回非零值
  - 否则返回零

2. inet_addr() 转换 IP 地址

```c
in_addr_t inet_addr(const char *cp)
```

- 功能：
  将网络主机地址 cp 从 IPv4 十进制点分地址转换成网络字节序的二进制数据
- 参数：
  cp:点分形式的 IP 地址(字符串)
- 返回值：
  如果输入无效 INADDR_NONE(通常是-1)会返回
  返回结果是 32 位整数地址(内部包含了字节序的转换，默认是大端模式)
- 特点：
  1. 仅适用于 IPv4
  2. 出错时会返回-1 而-1 对应 255.255.255.255 是有效的地址
  3. 此函数不能用于 255.255.255.255 的转换
  4. 不建议使用该函数，返回值

3. inet_pton()

```c
#include<arpa/inet.h>
int inet_pton(int af, const char *src, void *dst);
```

- 功能：将字符串 src 转换为指定地址族（af）的网络地址结构，然后将网络地址结构复制到 dst 中
- 特点：
  1. 适应于 IPV4 和 IPV6
  2. 能正确处理 255.255.255.255 的转换问题
- 参数：
  1. af：地址协议族（AF_INET 或 AF_INET6)必须是 AF_INET 或 AF_INET6
  2. src: 是一个指针（填写点分形式的 IP 地址（主要指 IPV4））
  3. dst：转换的结果给到 dst 以网络字节顺序写入
- 返回值：
  1. 成功转换时函数返回 1（网络地址验证成功）
  2. 如果 src 不包含代表指定地址系列中有效网络地址的字符串，则返回 0
  3. 如果 af 不包含有效的地址族，则返回 -1 并将 errno 设为 EAFNOSUPPORT

4. inet_ntop()

```c
#include <arpa/inet.h>
const char *inet_ntop(int af, const void *src,char *dst, socklen_t size);
```

- 功能：将 IPV4/IPV6 的网络字节序地址，变成本地的字符串形式的 IP 地址
- 特点：能正确的处理（-1）255.255.255.255 的转换问题
- 参数：
  1.  af: 地址协议族（AF_INET 或 AF_INET6）
  2.  src:是一个指针（32 位的网络字节序 IP 地址）
  3.  dst: 转换的结果为点分形式的 IP 地址[主要指 IPV4]
  4.  size：src 的长度
- 返回值
  - 成功将返回一个非空指针到 dst
  - 如果有错误将返回 NULL，并且 erron 将指示错误

- 在 OSI 模型中，主要位于会话层和传输层之间

## TCP 通信原理


### TCP的含义

TCP：Transfer Control protocol 传输控制协议 
提供面向连接的，一对一的可靠数据传输协议（双方要构建连接）
用于两个位于不同主机上的应用间的网络通信 

![alt text](TCP的含义.png)

### TCP 面向字节流

TCP 中的“流”(stream) 指的是流入或流出进程的字节序列

虽然应用程序和 TCP 的交互是一次一个数据块，但 TCP 把应用程序交下来的数据看成仅仅是一连串无结构的字节流    

![](TCP字节流.png)

TCP 不关心应用进程一次把多长的报文发送到 TCP 缓存

TCP 根据对方给出的窗口值和当前网络拥塞程度来决定一个报文段应包含多少个字节，形成 TCP 报文段

### TCP是如何确保传输可靠性的？

可靠传输的要求：

1. 传输信道不产生差错  （要使用一种方法保证A和B通信不会发送到别人那里）
   
2. 不管发送方以多快的速度发送数据，接收方总是来得及处理收到的数据（要使用一种方法来协调两者传输过程）

> `停止传输协议`(确认重发机制)

1. 每发送完一个分组就停止发送，等待对方的确认，在收到确认后再发送下一个分组

2. 全双工通信的双方既是发送方也是接收方


假设仅考虑 A 发送数据，而 B 接收数据并发送确认。因此 A 叫做发送方，而 B 叫做接收方

### TCP建立连接(三次握手)

## TCP 通信的流程

![alt text](TCP通信流程.png)

+ 通信之前，客户和服务器先创建套接字
+ 服务器端调用 bind，把端口号和本地 IP 地址填写到已创建的套接字中
+ 服务器端调用 listen（收听），把套接字设置为被动方式，以便随时接受客户的服务请求。UDP 服务器不使用 listen 系统调用
+ 服务器端调用 accept（接受），以便把远地客户进程发来的连接请求提取出来。UDP 服务器不使用 accept 系统调用
+ 客户进程调用 connect，以便和远地服务器建立连接（这就是主动打开）
+ 客户和服务器在 TCP 连接上使用 write传送数据，使用 read接收数据
+ 客户或服务器通信结束，调用 close 释放连接和撤销套接字
  
## 网络编程相关 API  

+ socket()　创建套接字  
+ bind()　绑定本机地址和端口  
+ connect()　建立连接  
+ listen()　设置监听套接字,把主动套接字变成动套接字  
+ accept()　接受 TCP 连接 ,阻塞等待客户端连接请求  
+ recv(), read(), recvfrom()　数据接收  
+ send(), write(),sendto()　数据发送  
+ close(), shutdown()　关闭套接字

## 3. listen()函数

```C
#include<sys/types.h>
#include<sys/socket.h>
int listen (int sockfd, int backlog);
```

- 功能：
  监听模式
- 参数：
  - sockfd:监听连接的套接字
  - backlog:  
    指定了正在等待连接的最大队列长度  
    它的作用在于处理可能同时出现的几个连接请求。同时允许几路客户端的服务器进行正在连接过程（正在三次握手）  
    DoS(拒绝服务)攻击即利用了这个原理，非法的连接占用了全部的连接数，造成正常的连接请求被拒绝  
    一般填 5， 测试得知,ARM 最大为 8
- 返回值： 成功返回 0 或 -1

内核中服务器的套接字 fd 会维护 2 个链表：

- 正在三次握手的的客户端链表(数量=2\*backlog+1)
- 已经建立好连接的客户端链表（已经完成３次握手分配好了**newfd**）

比如：listen(fd, 5); //表示系统允许 11(=2\*5+1)个客户端同时进行三次握手

完成 listen()调用后，socket 变成了监听 socket(listening socket).

````c
//server.c
#include "net.h"
int main()
{

/*1.创建socket fd*/
    int fd = -1;
    if((fd = socket(AF_INET, SOCK_STREAM, 0)) < 0)
    {
        perror("socket");
        exit(1);
    }

/*2.绑定*/
/*2.1填充struct_sockaddr_in 结构体变量*/
    struct sockaddr_in sin;
    bzero(&sin,sizeof(sin));                     //清空结构体内容
    sin.sin_family = AF_INET;                    //使用IPv4地址族
    sin.sin_port = htons(SERV_PORT);             //host to network 将本地字节序转成网络字节序
    //两种地址填充方式
    //sin.sin_addr.s_addr = inet_addr(SERV_IP_ADDR);  //inet_addr函数不建议使用
    // 优化1. 让服务器程序能绑定在任意IP上
#if 1
    sin.sin_addr.s_addr = htonl(INADDR_ANY);
#else
    if((inet_pton(AF_INET, SERV_IP_ADDR, (void *)&sin.sin_addr.s_addr))!=1)  // 转换+填充
    {   //该函数有可能失败
        perror("inet_pton");
        exit(1);   
    }
#endif
/*2.2 绑定*/
    if(bind(fd, (struct sockaddr *)&sin, sizeof(sin)) < 0)
    {
        perror("bind");
        exit(1);
    }

/*3. 调用listen（）把主动套接字变成被动套接字*/
    if(listen(fd, BACKLOG) < 0)
    {
        perror("listen");
        exit(1);
    }


    int newfd = -1;
/*4. 阻塞等待客户端连接请求*/
#if 0
    newfd = accept(fd, NULL, NULL);
    if(newfd < 0)
    {
        perror("accept");
        exit(1);
    }
#else
    /*优化2 通过程序获取刚建立的socket的客户端的IP地址和端口号*/
    struct sockaddr_in cin;  //新建一个结构体数据用来接收对方的地址
    socklen_t addrlen = sizeof(cin);      // 套接字的长度结构类型
    if( (newfd = accept(fd, &cin, &addrlen) ) < 0)
    {
        perror("accept");
        exit(1);
    }
    char ipv4_addr[16];    //用来存储地址
    if(inet_ntop(AF_INET, (void *)&cin.sin_addr.s_addr, ipv4_addr, sizeof(cin)))  //将网络字节序转换成本地
    {
        perror("inet_ntop");
        exit(1);
    }
    printf("Client(%s:%d) is connected!\n", ipv4_addr, htons(cin.sin_port));

    /*5. 读写*/
    int ret = -1;
    char buf[bufSIZ];
    while(1)
    {
        bzero(buf, bufSIZ);
        
        do
        {
            ret = read(newfd, buf, bufSIZ-1);
        }
        while(ret < 0 && EINTR == errno);

        if (ret < 0)
        {
            perror("read");
            exit(1);
        }
        if(!ret)                                //对方已经关闭
        {
            break;
        }
        printf("Receive data:%S\n", buf);   

        if(strcasecmp(buf, QUIT_STR, strlen(QUIT_STR))) //用户输入了quit字符
        {
            printf("Client is exiting!\n");
            break;
        }
    }
    close(newfd);
    close(fd);
    return 0;
}
````

````c
//client.c
#include "net.h"

int main(void)
{
	int fd = -1;	
/* 1. 创建socket fd*/
	if ((fd = socket(AF_INET, SOCK_STREAM, 0)) < 0)
	{
		perror("socket");
		exit(1);
	}

/*2.连接服务器 */
/*2.1 填充struct sockaddr_in结构体变量 */	
	struct sockaddr_in sin;
	bzero(&sin, sizeof(sin));     //清零

	sin.sin_family = AF_INET;   //网络协议族
	sin.sin_port = htons(SERV_PORT); //将主机字节序转成网络字节序

#if 0
	sin.sin_addr.s_addr = inet_addr(SERV_IP_ADDR);  //
#else
	if(inet_pton(AF_INET, SERV_IP_ADDR, (void *)&sin.sin_addr) != 1)    //转换IP地址
	{
		perror("inet_pton");
		exit(1);
	}
#endif
//连接到套接字fd上
	if(connect(fd, (struct sockaddr *)&sin, sizeof(sin)) < 0)           //连接
	{
		perror("connect");
		exit(1);
	}

	printf("Client staring...OK!\n");
/*3. 读写数据 */
	char buf[BUFSIZ];
	int ret = -1;
	while (1)
	{
		bzero(buf, BUFSIZ);
		if (fgets(buf, BUFSIZ - 1, stdin) == NULL)  //从标准输入读取一行数据
		{
			continue;
		}
		do
		{
			ret = write(fd, buf, strlen(buf));
		} 
		while (ret < 0 && EINTR == errno);

		if ( !strncasecmp(buf, QUIT_STR, strlen(QUIT_STR))) 
		{ //用户输入了quit字符
			printf("Client is exiting!\n");
			break;
		}
	}
/*4.关闭套接字 */
	close(fd);
}
````

````c
//server
#include <stdio.h>
#include <stdlib.h>
#include <unistd.h>
#include <sys/socket.h>
#include <sys/types.h>
#include <arpa/inet.h>
#include <string.h>
 
#define BACKLOG 5
 
int main(int argc, char *argv[])
{
	int fd, newfd, ret;  
	char buf[BUFSIZ] = {};                           //传输的数据
	struct sockaddr_in addr;                        //结构体负责存储iP地址和端口号

	if (argc < 3)                                   //提示传入的参数个数
    {
		fprintf(stderr, "%s<addr><port>\n", argv[0]);
		exit(0);
	}
 
	/*创建套接字 */
	fd = socket(AF_INET, SOCK_STREAM, 0);     
	if (fd < 0)
    {
		perror("socket");
		exit(0);
	}
    //填充 addr 结构体
	addr.sin_family = AF_INET;
	addr.sin_port = htons(atoi(argv[2]));  					//端口号转化
	if (inet_aton(argv[1], &addr.sin_addr) == 0) 			//ip地址转化
    {
		fprintf(stderr, "Invalid address\n");  				//函数不在perror中留下信息
		exit(EXIT_FAILURE);
	}
 
	/*绑定通信结构体*/
	if(bind(fd, (struct sockaddr *)&addr, sizeof(addr)) == -1)
    {
		perror("bind");
		exit(0);
	}
 
	/*设置套接字为监听模式*/
	if (listen(fd, BACKLOG) == -1) 
    {
		perror("listen");
		exit(0);
	}
 
 
	/*接受客户端的连接请求,生成新的用于和客户端通信的套接字*/
	newfd = accept(fd, NULL, NULL);
	if (newfd < 0) 
    {
		perror("newfd");
		exit(0);
	}
 
	while(1) 
    {
		memset(buf, 0, BUFSIZ);
		ret = read(newfd, buf, BUFSIZ);
		if(ret < 0) 
        {
			perror("read");
			exit(0);
		}else if(ret == 0)
			break;
		else
			printf("buf = %s\n", buf);
	}
 
	close(newfd);
	close(fd);
 
	return 0;
 
}
````

````c
#include <pthread.h>
#include <signal.h>
#include "net.h"


void cli_data_handle (void *arg);

void sig_child_handle(int signo)
{
	if (SIGCHLD == signo) 
	{
		waitpid(-1, NULL,  WNOHANG);
	}
}
int main (void)
{

	int fd = -1;
	struct sockaddr_in sin;

	signal(SIGCHLD, sig_child_handle);      //使用信号以防止子进程变成僵尸进程

	/* 1. 创建socket fd */
	if ((fd = socket (AF_INET, SOCK_STREAM, 0)) < 0) 
	{
		perror ("socket");
		exit (1);
	}

	int b_reuse = 1;
	setsockopt (fd, SOL_SOCKET, SO_REUSEADDR, &b_reuse, sizeof (int));

	/*2.1 填充struct sockaddr_in结构体变量 */
	bzero (&sin, sizeof (sin));
	sin.sin_family = AF_INET;
	sin.sin_port = htons (SERV_PORT);	//网络字节序的端口号

#if 1
	sin.sin_addr.s_addr = htonl (INADDR_ANY);
#else
	if (inet_pton (AF_INET, SERV_IP_ADDR, (void *) &sin.sin_addr) != 1) 
	{
		perror ("inet_pton");
		exit (1);
	}
#endif

/*2.2 绑定 */
	if (bind (fd, (struct sockaddr *) &sin, sizeof (sin)) < 0) 
	{
		perror ("bind");
		exit (1);
	}

/*3. 调用listen()把主动套接字变成被动套接字 */
	if (listen (fd, BACKLOG) < 0) 
	{
		perror ("listen");
		exit (1);
	}
	printf ("Server starting....OK!\n");
	int newfd = -1;
/*4. 阻塞等待客户端连接请求 */
	struct sockaddr_in cin;
	socklen_t addrlen = sizeof (cin);
	while (1)
	{
		pid_t pid = -1;
		if((newfd = accept(fd, (struct sockaddr *) &cin, &addrlen)) < 0) 
		{
			perror ("accept");
			break;
		}
		/*创建一个子进程用于处理已建立连接的客户的交互数据*/
		if ((pid = fork()) < 0) 
		{
			perror("fork");
			break;
		}

		if (0 == pid) 
		{ //子进程中
			close(fd);                  //子进程不需要fd资源
			char ipv4_addr[16];
			if (!inet_ntop (AF_INET, (void *) &cin.sin_addr, ipv4_addr, sizeof (cin))) 
			{
				perror ("inet_ntop");
				exit (1);
			}
			printf ("Clinet(%s:%d) is connected!\n", ipv4_addr, ntohs(cin.sin_port));
			cli_data_handle(&newfd);
			return 0;
		} 
		else 
		{ //实际上此处 pid >0, 父进程中
			close(newfd);                     //父进程不需要newfd
		}
	}
	close (fd);
	return 0;
}

void cli_data_handle (void *arg)
{
	int newfd = *(int *) arg;

	printf ("Child handling process: newfd =%d\n", newfd);

	//..和newfd进行数据读写
	int ret = -1;
	char buf[BUFSIZ];
	while (1) 
	{
		bzero (buf, BUFSIZ);
		do 
		{
			ret = read (newfd, buf, BUFSIZ - 1);
		} 
		while (ret < 0 && EINTR == errno);
		if (ret < 0) 
		{
			perror ("read");
			exit (1);
		}
		if (!ret) 
		{				//对方已经关闭
			break;
		}
		printf ("Receive data: %s\n", buf);

		if (!strncasecmp (buf, QUIT_STR, strlen (QUIT_STR))) {	//用户输入了quit字符
			printf ("Client(fd=%d) is exiting!\n", newfd);
			break;
		}
	}
	close (newfd);

}
````

假设有一台发送端主机A要发送一个数据包给接收端主机B，但是由于网络链路的限制，数据包的大小超过了链路的MTU（最大传输单元）。在这种情况下，主机A会将数据包进行分片，即将一个大的数据包分割成多个小的数据包进行传输。

下面是一个IP分片的示例过程：

假设主机A要发送的数据包大小为1500字节，而链路的MTU为1000字节。

主机A首先会将数据包分成多个片段，每个片段的大小为1000字节（除了最后一个片段可能会小于1000字节）。

主机A在IP首部的标识字段中设置一个唯一的标识符，以便接收端主机B能够将这些片段重新组装成原始的数据包。

主机A会为每个片段设置不同的偏移量字段，以指示片段在原始数据包中的位置。

主机A会设置片段的MF（More Fragments）位，以指示是否还有更多的片段需要传输。如果还有更多的片段，则MF位被设置为1，否则为0。

主机A给每个片段重新计算IP首部和校验和字段，并将这些片段依次发送给主机B。

接收端主机B收到这些片段后，会根据IP首部的标识符和偏移量字段来重新组装这些片段，以恢复原始的数据包。

主机B会检查MF位，如果还有更多的片段需要接收，则会等待接收其他片段，并继续组装。

当主机B接收到最后一个片段，并且MF位为0时，说明所有片段都已经接收完毕，主机B成功恢复了原始的数据包。

这个过程中，IP分片技术允许发送端主机将大的数据包分割成适合链路MTU的小片段进行传输，以确保数据包能够成功通过网络。然而，IP分片也会增加网络传输的复杂性和开销，因此在实际网络中，尽量避免使用IP分片是一个好的实践。
