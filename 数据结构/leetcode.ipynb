```python
struct ListNode
{
    int val;
    struct ListNode *next;
}
struct ListNode* addTwoNumbers(struct ListNode* l1, struct ListNode* l2) 
{
    struct ListNode *head = NULL;
    struct ListNode *tail = NULL;
    int carry = 0;  //初始进位为0
    while (l1 || l2) 
    {
        int n1 = l1 ? l1->val : 0;  //判断l1是否为空，如果非空，则取出其值，否则为0
        int n2 = l2 ? l2->val : 0;
        int sum = n1 + n2 + carry;
        if (!head)   //head为空，是第一个结点
        {
            head = tail = malloc(sizeof(struct ListNode));  //新建一个头结点
            tail->val = sum % 10;  //头结点的值，取个位
            tail->next = NULL;
        } 
        else  //不是第一个结点
        {
            tail->next = malloc(sizeof(struct ListNode));
            tail->next->val = sum % 10;  //新建结点，值取个位
            tail = tail->next;
            tail->next = NULL;
        }
        carry = sum / 10;  //进位，取十位
        if (l1)  //如果l1还有值，则继续取下一个结点 
        {
            l1 = l1->next;
        }
        if (l2) 
        {
            l2 = l2->next;
        }
    }
    if (carry > 0) //处理最后一位的进位
    {
        tail->next = malloc(sizeof(struct ListNode));
        tail->next->val = carry;
        tail->next->next = NULL;
    }
    return head;
}
```

```python
int lengthOfLongestSubstring(char * s)
{
    // 滑动窗口
    int left = 0;  // 左边界
    int right = 0;  // 右边界
    
    int max = 0;
    int i,j;
    int len = strlen(s);
    int haveSameChar = 0;

    for(i = 0; i < len ; i++)  //i 标记最左边的字符  
    {
        if(left <= right)  
        {   
            haveSameChar = 0;
            for(j = left; j < right ; j++)  //j 标记当前字符
            {
                if(s[j] == s[right])  // 判断当前字符是否与最左边的字符相同
                {
                    haveSameChar = 1;  // 相同则标记为1
                    break;  // 跳出循环
                }
            }
            if(haveSameChar)
            {
                left = j + 1;  //更新的左边界
            }
        }
        //统计最大的间距
        max = max < (right - left + 1) ? (right - left + 1): max;
        right++;  // 没找到相同字符，则更新右边界
    }
    return max;
}
```
